```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("weatherHistory.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc=True)

```

```python
df.head()
```

```python
df.set_index('Formatted Date',inplace =True)
df.head()
```

```python
#converting index type to dateindex so that we can apply resample function
df.index = pd.to_datetime(df.index)
```

```python
df.head(10)
```

```python
#Remove the row with data for year 2005
df.drop(df.index[0],inplace = True)
```

```python
df.head()
```

```python
# 'MS' means month start
df_new = df[['Apparent Temperature (C)','Humidity']]
df_new = df_new.resample('MS').mean()
```

```python
df_new.head()
```

```python
xmax=df_new['Apparent Temperature (C)'].max()
xmin=df_new['Apparent Temperature (C)'].min()
df_new.plot()

plt.show()
```

```python
#maximum value of apparent temperature in degree celsius
xmax
```

```python
#minimum value of apparent temperature in degree celsius
xmin
```

<h2><b>Observations:</b></h2>

1. The humidity remains almost constant across the decade from 2006-2016.

2. General observation that the apparent temperature starts increasing from January till mid year around June
 and starts falling there after till the end of the year and this occurs in the same pattern mostly every year.
 
3. Overall max Apparent Temperature of 24.52°C noted in month of July 2012.

4. overall max Apparent Temperature of -8.82°C noted in month of February 2012.Thus most extreme climate was found in the year 2012.

 *Many of the Finnish Meteorological Institute’s observation stations had the rainiest year within the past 50 years. Altogether 907 millimetres of precipitation was measured in Kaisaniemi, Helsinki for the year 2012.*
